# Uso de ciclos en Clojure

Definición de factorial, versión recursiva explícita.

```clojure
(defn !-v1
  [n]
  (if (zero? n)
    1
    (*' n (!-v1 (dec n)))))
```

```clojure
(!-v1 0)
```

```clojure
(!-v1 5)
```

```clojure
(!-v1 10)
```

```clojure
(!-v1 20)
```

```clojure
(!-v1 30)
```

```clojure
(!-v1 100)
```

```clojure
(!-v1 1000)
```

Definición de factorial, versión loop/recur.

```clojure
(defn !-v2
  [n]
  (loop [i 1
         result 1]
    (if (> i n)
      result
      (recur (inc i) (*' i result)))))
```

```clojure
(!-v2 0)
```

```clojure
(!-v2 5)
```

```clojure
(!-v2 1000)
```

Definición de factorial, versión API de secuencias.

```clojure
(range 5)
```

```clojure
(range 10)
```

```clojure
(range 10 20)
```

```clojure
(range 10 101 10)
```

```clojure
(defn !-v3
  [n]
  (reduce *' (range 1 (inc n))))
```

```clojure
(!-v3 5)
```

```clojure
(!-v3 1000)
```

```clojure
(*')
```

```clojure
(range 1 1)
```

```clojure
(!-v3 0)
```

# Operaciones básicas con secuencias

* map
* filter
* reduce

```clojure
(defn sqr
  [x]
  (* x x))
```

```clojure
(sqr 5)
```

```clojure
(map sqr '(4 -10 8 1 0 7))
```

```clojure
(map - '(3 5 -6 8))
```

```clojure
(filter even? '(4 6 7 10 14 15))
```

```clojure
(filter even? '(1 5 9 11 3))
```

```clojure
(reduce + '(4 5 8 -10))
```

```clojure
(list 5)
```

```clojure
(list 4 5)
```

```clojure
(reduce list '(1 2 3 4))
```

```clojure
(reduce list () '(1 2 3 4))
```

```clojure
(filter (fn [x] (< x 10)) '(1 5 10 17 8))
```

```clojure
(filter #(< % 10) '(1 5 10 17 8))
```
